```python
df = pd.read_csv('artigos_wit.csv')
```

Pré-Processamento

```python
# Detecção de Idiomas dos Títulos e Resumos
def detect_language(text):
    try:
        return detect(text)
    except:
        return 'unknown'
```

```python
import re
import spacy

stopwords = {"de", "a", "e", "o", "da", "do", "que", "em", "na", "uma", "um", "com", "as", "no", "os", "das"}

nlp = spacy.load("pt_core_news_sm")

def pre_processing(text):
    text = text.lower()
    text = re.sub(r"[,\-]", " ", text)
    doc = nlp(text)

    tokens = [
        token.text
        for token in doc
        if token.is_alpha and token.text not in stopwords
    ]

    return tokens
```

```python
df['idioma_titulo'] = df['titulo'].apply(detect_language)
df['idioma_abstract'] = df['abstract'].apply(detect_language)
```

```python
df['tokens'] = df['abstract'].apply(pre_processing)
```

```python
# Trecho para contar a frequência global dos tokens - Para ver a frequência inicial, deve-se comentar a lista de stopwords da função pre_processing

tokens_corpus = []

for lista_tokens in df['tokens']:
    tokens_corpus.extend(lista_tokens)

frequencia_global = Counter(tokens_corpus)

frequencia_global.most_common(50)
```
